<h3>Determine the RD curves for the MST</h3>
<p>To do this, we need to run two commands with the two different implementations.</p>
<ul>
    <li>python3 stereo_MST_coding_16.py --show_stats -t 25 -f ../data/VVV_InviernoNuclear.oga -e 6 -q 8</li>
    <li>python3 stereo_MST_coding_32.py --show_stats -t 25 -f ../data/VVV_InviernoNuclear.oga -e 6 -q 8</li>
</ul>
<p>The quantization parameter goes from 8 to 512, following the values of the powers of 2. Once all the necessary data is taken, we can represent it in a graphic.</p>

```python
import matplotlib.pyplot as plt

x1=[797.12, 760.43, 690.32, 565.98, 435.56, 299.85, 202.31]
y1=[0.0021, 0.0047, 0.0097, 0.0192, 0.037, 0.0803, 0.151]

x2=[826.32, 790.21, 764.89, 696.64, 578.98, 420.34, 305.87]
y2=[0.0012, 0.0020, 0.0050, 0.0093, 0.0193, 0.0354, 0.071]

plt.plot(x1, y1, label="16 bits case")
plt.plot(x2, y2, label="32 bits case")

plt.xlabel("kbps")
plt.ylabel("RMSE")
plt.title("RD Curves of the Stereo MST Coding")

plt.legend()
plt.show()
```

<p>We can also do a comparison between all the graphics.</p>

```python
import matplotlib.pyplot as plt

x1=[797.12, 760.43, 690.32, 565.98, 435.56, 299.85, 202.31]
y1=[0.0021, 0.0047, 0.0097, 0.0192, 0.037, 0.0803, 0.151]

x2=[826.32, 790.21, 764.89, 696.64, 578.98, 420.34, 305.87]
y2=[0.0012, 0.0020, 0.0050, 0.0093, 0.0193, 0.0354, 0.071]

x3=[527.1, 440.23, 361.21, 300.73, 269.12, 232.89, 172.92]
y3=[0.0012, 0.0018, 0.0033, 0.0056, 0.01, 0.0192, 0.0342]

x4=[531.34, 446.76, 356.23, 300.41, 256.56, 231.19, 172.43]
y4=[0.0015, 0.0024, 0.0035, 0.0059, 0.011, 0.0125, 0.04]

plt.plot(x1, y1, label="MST 16 bits case")
plt.plot(x2, y2, label="MST bits case")
plt.plot(x3, y3, label="NO Overlapped DWT")
plt.plot(x4, y4, label="Overlapped DWT")

plt.xlabel("kbps")
plt.ylabel("RMSE")
plt.title("RD Curves of the different codings")

plt.legend()
plt.show()
```

<p>As we can see in the last graphic, the RMSE is higher in the MST Coding, while the DWT Coding has a lower one. This means that the error between the predicted values and the real ones are lower.</p>

<h4>Does the computation of the DWT using chunk overlapping increase the latency of the whole system? And the jitter? In which amount?</h4>
<p>The latency does increase, but the jitter not. This is because the chunk overlapping increase the time that takes the process to send the packet and receive the acknowledge, but not the fluctuation of the latencies.</p>
<h4>Which other transform(s) are used in audio encoding systems (such as MP3) for exploiting the temporal redundancy? Enumerate the systems and the used transform(s).</h4>
<ul>
    <li>Discrete cosine transform. A discrete cosine transform (DCT) expresses a finite sequence of data points in terms of a sum of cosine functions oscillating at different frequencies.</li>
    <li>Discrete sine transform. It is a Fourier-related transform similar to the discrete Fourier transform (DFT), but using a purely real matrix. It is equivalent to the imaginary parts of a DFT of roughly twice the length, operating on real data with odd symmetry (since the Fourier transform of a real and odd function is imaginary and odd), where in some variants the input and/or output data are shifted by half a sample.</li>
    <li>The Daubechies wavelets. Based on the work of Ingrid Daubechies, are a family of orthogonal wavelets defining a discrete wavelet transform and characterized by a maximal number of vanishing moments for some given support. With each wavelet type of this class, there is a scaling function (called the father wavelet) which generates an orthogonal multiresolution analysis.</li>
</ul>
